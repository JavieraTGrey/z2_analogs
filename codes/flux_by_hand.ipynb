```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from astropy.io import fits
```

```python
DIR = '/Users/javieratoro/Desktop/proyecto 2024-2/'
J0023_data = fits.open(DIR + 'codes/J0023.fits')[1].data
lines = pd.read_csv(DIR + 'CSV_files/emission_lines.csv')

wave, flux = J0023_data['wave'], J0023_data['flux']
```

```python
fig = plt.figure()
plt.plot(wave, flux)
plt.show()
```

```python
rf_wave = ( (3729.875447997208 + 3727.091727) / 2 ) * (1 + 0.053106597326384564)
mask = (wave > rf_wave - 4) & (wave < rf_wave + 4)
obs_wave, obs_flux = wave[mask], flux[mask]

fig = plt.figure()
plt.plot(obs_wave, obs_flux)
plt.show()
```
