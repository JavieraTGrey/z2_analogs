```python
def mag_to_flux_sinh(mag, b):
    '''
    Takes an SDSS magnitude and the softening parameter b to calculate the
    flux density in mJy.
    '''
    f_f0 = 2 * b * unumpy.sinh(- np.log(b) - (mag * np.log(10) / 2.5 ))
    S = 3631 * 1e3 * f_f0
    return S

def mag_to_flux_AB(mag):
    """
    Convert AB magnitudes to flux densities in microjansky (mJy).
    """
    fluxes = np.where(mag == 9999.0, None, 10 ** ((23.9 - mag) / 2.5))/1000
    return fluxes
```

```python
# Prepare columns
ID = np.array(mage['ID'])
redshift = mage['z']

# Create dictionary

cigale_input = {'id': ID, 'redshift': redshift}

# SDSS bands and softening parameters
sdss_bands = ['u', 'g', 'r', 'i', 'z']
b = np.array([1.4, 0.9, 1.2, 1.8, 7.4]) * 1e-10


if __name__ == '__main__':

    # Create table
    for band in sdss_bands:
        # Gather the data
        band_mag = unumpy.uarray(mage['modelMag_'+band],
                                 mage['modelMagErr_'+band])

        # Corrections according to Flux Calibration from SDSS
        if band == 'u':
            band_mag = band_mag - 0.04

        if band == 'z':
            band_mag = band_mag + 0.02

        b_band = b[np.argmax(sdss_bands == band)]

        band_flux = mag_to_flux_sinh(band_mag, b_band)

        cigale_input[band+'_prime'] = unumpy.nominal_values(band_flux)
        cigale_input[band+'_prime_err'] = unumpy.std_devs(band_flux)

    # Create table and write to input.txt
    final_input = Table(cigale_input)
    savepath = homedir + 'sed_fitting/'
    final_input.write(savepath + 'mage/cigale_mage_input.txt', format='ascii', overwrite=True)
    print('Cigale input file saved to', savepath)
```

```python
ID = np.array(cecilia['Object'])
redshift = cecilia['Redshift']

# Create dictionary

cigale_input = {'id': ID, 'redshift': redshift}

# SDSS bands and softening parameters
cecilia_bands = [col for col in cecilia.columns[3:-1] if '_err' not in col]

names = ['keck.lris.g', 'keck.lris.rs', 'keck.mosfire.j', 'keck.mosfire.h',
         'keck.mosfire.ks', 'spitzer.irac.ch1', 'spitzer.irac.ch2',
         'spitzer.irac.ch3', 'hst.wfc3.F110W', 'hst.wfc3.F140W',
         'hst.wfc3.F160W']

if __name__ == '__main__':

    # Create table
    for i in range(0, len(names)):


        # Gather the data
        band = cecilia_bands[i]
        band_mag = unumpy.uarray(cecilia[band].values,
                                 cecilia[band + '_err'].values)

        # Corrections according to Flux Calibration from SDSS

        band_flux = mag_to_flux_AB(band_mag)
        # cigale_input[names[i]] = np.where(band_flux == 0.0, 0.0, unumpy.nominal_values(band_flux))
        # cigale_input[names[i] + '_err'] = np.where(band_flux == 0.0, 1.0, unumpy.std_devs(band_flux))
        cigale_input[names[i]] = np.where(band_flux == 0.0, None, unumpy.nominal_values(band_flux))
        cigale_input[names[i] + '_err'] = np.where(band_flux == 0.0, None, unumpy.std_devs(band_flux))

    #         # Create table and write to input.txt
    del cigale_input['spitzer.irac.ch1']
    del cigale_input['spitzer.irac.ch1_err']
    del cigale_input['spitzer.irac.ch3']
    del cigale_input['spitzer.irac.ch3_err']

    # final_input = pd.DataFrame(cigale_input)
    # savepath = homedir + 'sed_fitting/'
    # final_input.to_csv(savepath + 'cecilia/cigale_cecilia_input.txt', sep=' ', index=False)
    final_input = Table(cigale_input)
    savepath = homedir + 'sed_fitting/'
    final_input.write(savepath + 'cecilia/cigale_cecilia_input.txt', format='ascii', overwrite=True)
    print('Cigale input file saved to', savepath)
```
